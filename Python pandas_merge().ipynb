Merge(): connects columns or indexes in DataFrame based on one or more keys

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'ID':[1,2,3,4,5],'Class' :[8,9,10,11,12]})
df1
```

```python
df2 = pd.DataFrame({'ID':[1,2,3,4],'Section' :['A','B','C','D']})
df2
```

```python
pd.merge(df1,df2, on='ID')
```

```python
pd.merge(df2,df1, on='ID')
```

```python
# by default how = 'inner'
pd.merge(df1,df2, on='ID', how='inner')
```

```python
pd.merge(df1,df2,on='ID',how='left')
```

```python
pd.merge(df1,df2,on='ID',how='right')
```

```python
pd.merge(df1,df2,on='ID',how='outer')
```

```python
# To know in which df the value is from:indicater = True
pd.merge(df1,df2,on='ID', how='outer', indicator=True)
```

```python
df1.merge(df2,on='ID')
```

```python
df1.merge(df2,on='ID',how='left')
```
