```python
df1 = pd.DataFrame({
    'A' : [1,2,3,4,5],
    'B' : [10,20,30,40,50]
}, index = ['i1','i2','i3', 0,1])

df2 = pd.DataFrame({
    'A': [4,5,6],
    'B': [40,50,60]
})
```

```python
# It throws error because of same column names
df1.join(df2)
```

```python
# Use parameter L_suffix or R_suffix  to avoid such error
df1.join(df2, lsuffix = '_df1')
```

```python
df1.join(df2, rsuffix = '_df2')
```

```python
# how = 'left' by default
df1.join(df2, lsuffix='_df1' ,rsuffix = '_df2')
```
