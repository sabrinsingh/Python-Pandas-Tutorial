# Handling Missing Values in Python pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('Fortune_10.csv')
df
```

We can see multiple NaN values. Also in industry field, there are values like 'no values' and 'not available' which should also be considered as Nan

```python
df= pd.read_csv('Fortune_10.csv', na_values = ['no values', 'not available'])
df
```

```python
df= pd.read_csv('Fortune_10.csv', na_values = {'Industry': ['no values']})
df
```

```python
df= pd.read_csv('Fortune_10.csv',keep_default_na =False)
df
```

```python
df= pd.read_csv('Fortune_10.csv',keep_default_na = True)
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
#opposite of isnull()
df.notnull()
```

```python
# deleting all the rows/column with NULL (NaN) values
df.dropna()
```

```python
df
```

```python
help(df.dropna)
```

```python
#delete the columns (axis = 1) with NaN values
df.dropna(axis = 1)
```

```python
#by defult input
df.dropna(how = 'any', axis=0)
```

```python
df.dropna(how = 'any' , axis = 1)
```

```python
#delete the rows if all the values in row in NaN
df.dropna(how = 'all')
```

```python
#thresh is the short for threshold. This means displaying the value with atleast the number of threshold value
# e.g thresh = 1 means it shows the row/col (axis=0/axis=1) with atleast 1 not null valu
df.dropna(thresh=2)
```

```python
df.dropna(thresh=7)
```

```python
# inplace = False by default
df.dropna(inplace = True)
```
