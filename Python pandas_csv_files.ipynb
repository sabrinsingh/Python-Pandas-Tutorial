### If need to check the particular number of rows

```python
df = pd.read_csv('C:\\Users\\heart\\Downloads\\Fortune_10.csv', nrows = 5)
df
```

### If only certain columns are needed from the csv file 

```python
df = pd.read_csv('C:\\Users\\heart\\Downloads\\Fortune_10.csv', usecols = [2,4,7])
df
```

### To Skip first rows

```python
df = pd.read_csv('C:\\Users\\heart\\Downloads\\Fortune_10.csv', skiprows = 2)
df
```

### To skip the specific rows

```python
# Lets say skip 2st and 4th row
df = pd.read_csv('C:\\Users\\heart\\Downloads\\Fortune_10.csv', skiprows = [2,3])
df
```

### To change the index from one of the columns of the file

```python
df = pd.read_csv('C:\\Users\\heart\\Downloads\\Fortune_10.csv', index_col = 'ID' )
df
```

```python
df.reset_index()
```
