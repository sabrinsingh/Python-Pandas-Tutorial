## Pandas Pivot Table
### The levels in the pivot table will be stored in MultiIndex objects(hierarchical indexes) on the index and columns of the result DataFrame

df.pivot_table()

```python
import pandas as pd
```

```python
movies_df= pd.read_csv('movies_metadata.csv',low_memory=False)

movies_df
```

```python
df = pd.read_csv('movies_metadata.csv', usecols = [1,7,5,6]).head(10)
df
```

```python
# adding the new column after eliminating the first character of the string
df['Worldwide Gross(in$)'] = df["Worldwide Gross"].str[1:]
```

```python
#deleting the column
del df["Worldwide Gross"]
```

```python
df.info()
```

```python
# Changing the datatype
df[['Worldwide Gross(in$)']] = df[['Worldwide Gross(in$)']].apply(pd.to_numeric)
```

```python
df.info()
```

```python
df
```

```python
df.pivot_table(index='Genre')
```

```python
df.pivot_table(index = 'Genre', aggfunc='sum')
```

```python
df.pivot_table(index= 'Genre', columns = 'Year')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', aggfunc='count')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', aggfunc='max')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', fill_value = 'none')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', fill_value = 0)
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', fill_value = 0 , margins = True)
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', aggfunc='sum', fill_value = 0 , margins = True)
```
