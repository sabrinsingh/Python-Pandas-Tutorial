Pandas interpolate() function is basically used to fill NaN values in dataframe or series.

    Syntax: Dataframe.interpolate()
            Series.interpolate()
            
    Parameters: method inplace axis limit limit_direction limit_area


```python
import pandas as pd
```

```python
df = pd.read_csv('Fortune_10.csv')
df
```

```python
#interpolate() function can be used to fill only the numerical value
df.interpolate()
```

```python
df = pd.read_csv('Fortune_10.csv', index_col= 'Profit')
df
```

```python
df.interpolate(method = 'index')
```

```python
df
```

```python
df.interpolate(method = 'nearest')
```

```python
df.interpolate(method = 'polynomial',order =2)
```
