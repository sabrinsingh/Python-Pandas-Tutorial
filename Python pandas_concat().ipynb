```python
import pandas as pd
```

```python
sr1= pd.Series([0,1,2,3])
sr1
```

```python
sr2= pd.Series([4,5,6,7,8])
sr2
```

```python
pd.concat([sr1,sr2])
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
              'Name':['A','B','C','D'],
              'Class' : [7,8,9,10]
                })
df1
```

```python
df2=pd.DataFrame({'ID':[11,12,13,14],
              'Name':['AB','BB','CB','DB'],
              'Class' : [17,18,19,20]
                })
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2,df2])
```

```python
#axis = 0 by default
pd.concat([df1,df2], axis=1 )
```

```python
# ignore_index = False by default.ignore
pd.concat([df1,df2],ignore_index =True)
```
