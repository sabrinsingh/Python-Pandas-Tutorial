```python
import pandas as pd
```

```python
sr1= pd.Series([0,1,2,3])
sr1
```

```python
sr2= pd.Series([4,5,6,7,8])
sr2
```

```python
pd.concat([sr1,sr2])
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
              'Name':['A','B','C','D'],
              'Class' : [7,8,9,10]
                })
df1
```

```python
df2=pd.DataFrame({'ID':[11,12,13,14],
              'Name':['AB','BB','CB','DB'],
              'Class' : [17,18,19,20]
                })
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2,df2])
```

```python
#axis = 0 by default
pd.concat([df1,df2], axis=1 )
```

```python
# ignore_index = False by default.ignore_index = True will concatenating the index and reset the index
pd.concat([df1,df2],ignore_index =True)
```

```python
df3=pd.DataFrame({'ID':[1,2],
              'Name':['A','B'],
              'Class' : [7,8]
                })
df3
```

```python
pd.concat([df2,df3], ignore_index=True)
```

```python
pd.concat([df2,df3], axis = 1)
```

```python
pd.concat([df2,df3], axis = 1, join = 'inner')
```

```python
pd.concat([df2,df3], axis = 1, join = 'outer')
```

```python
pd.concat([df2,df3],axis=1, join_axes=[df2.index])
```
